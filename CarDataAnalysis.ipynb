```python
import pandas as pd
import numpy as np
```

```python
cleaned= pd.read_csv("cleaned.csv")
cleaned

cleaned.dropna(subset=['Ex-Showroom_Price', 'Displacement'], inplace=True)
cleaned['Ex-Showroom_Price'] = cleaned['Ex-Showroom_Price'].str.replace('[^\d]', '', regex=True).astype(int)
cleaned['Displacement'] = cleaned['Displacement'].str.replace('[^\d]', '', regex=True).astype(int)
```

```python
import numpy as np

# Fill NaN values with 0
cleaned.fillna(0, inplace=True)

# Function to handle conversion of mileage
def convert_mileage(x):
    parts = str(x).split('-')
    if len(parts) > 1:
        return int(float(parts[0]))
    else:
        return int(float(''.join(filter(str.isdigit, str(x)))))


# Convert ARAI_Certified_Mileage to integer
cleaned['ARAI_Certified_Mileage'] = cleaned['ARAI_Certified_Mileage'].apply(convert_mileage)

# Remove text and convert Kerb_Weight to integer
cleaned['Kerb_Weight'] = cleaned['Kerb_Weight'].apply(lambda x: int(str(x).split()[0]))

# Keep only the numerical values in Power column
cleaned['Power'] = cleaned['Power'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if str(x).isdigit() else int(''.join(filter(str.isdigit, str(x.split()[0])))))

# Keep only the numerical values in Torque column
cleaned['Torque'] = cleaned['Torque'].apply(lambda x: int(''.join(filter(str.isdigit, str(x)))) if str(x).isdigit() else int(''.join(filter(str.isdigit, str(x.split()[0])))))

cleaned['Boot_Space'] = cleaned['Boot_Space'].apply(lambda x: int(str(x)[:3]))# Display the cleaned dataframe
print(cleaned)
cleaned['ARAI_Certified_Mileage'] = cleaned['ARAI_Certified_Mileage'] / 10
cleaned['ARAI_Certified_Mileage'] = cleaned['ARAI_Certified_Mileage'].round(1)
cleaned.to_csv("Cleaned.csv")
```
